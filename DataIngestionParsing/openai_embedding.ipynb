```python
from langchain_openai import OpenAIEmbeddings


embeddings = OpenAIEmbeddings(
    base_url="",
    model="text-embedding-3-large"
)
```

```python
## create your first embeddings
input_text = "Hello, I am learning about embeddings!"

single_embedding = embeddings.embed_query(input_text)
print(f"Text: {input_text}")
print(f"Embedding Length : {len(single_embedding)}")
print(single_embedding)
```

```python
# create embedding for list of documents
sentences = [
    "The quick brown fox jumps over the lazy dog.",
    "Data science drives decisions in modern organizations.",
    "Machine learning models improve with more data.",
    "Embedding vectors capture semantic meaning of text.",
    "Python is a versatile programming language.",
    "AI can assist humans in complex tasks.",
    "Cloud computing provides scalable resources.",
    "Collaborative teams build better software.",
    "Documentation helps users understand code.",
    "Testing ensures reliability of applications."
]

multiple_embedding = embeddings.embed_documents(sentences)
print("Embedded sentences : ")
for idx in range(len(sentences)):
    print(f"{sentences[idx]} : {multiple_embedding[idx]}")
```

```python
import numpy as np

def cosine_similarity(vec1, vec2):
    """
    cosine similarity measures the angle between two vectors.
    - result close to 1: very similar
    - result close to 0: not related
    - result close to -1: opposite meanings
    """

    dot_product = np.dot(vec1, vec2)
    norm_a = np.linalg.norm(vec1)
    norm_b = np.linalg.norm(vec2)
    return dot_product/(norm_a*norm_b)
```

```python
for idx in range(len(multiple_embedding)):
    for idx2 in range(idx+1, len(multiple_embedding)):
        similarity = cosine_similarity(multiple_embedding[idx], multiple_embedding[idx2])
        
        print(f"'{sentences[idx]}' vs '{sentences[idx2]}'")
        print(f"similarity: {similarity:.3f}\n")
```

```python
sentences = [
    "The cat is sleeping on the sofa in the living room.",
    "A dog is running across the park chasing a ball.",
    "The chef is preparing pasta in the kitchen.",
    "A programmer is writing Python code on a laptop.",
]


query = "Someone is coding a Python program on their computer."
embedded_sentences = embeddings.embed_documents(sentences)
embedded_question = embeddings.embed_query(query)
for idx in range(len(embedded_sentences)):
    similarity = cosine_similarity(embedded_sentences[idx], embedded_question)
    print(f"'{sentences[idx]}' vs '{query}' : {similarity:.3f}")
```

```python
def semantic_search(query, documents, embeddings_models: OpenAIEmbeddings, top_k):
    """Simple semantic search implementation"""

    query_embedding = embeddings_models.embed_query(query)
    documents_embeddings = embeddings_models.embed_documents(documents)

    similarities = []
    for idx, document_embedding in enumerate(documents_embeddings):
        similarity = cosine_similarity(document_embedding, query_embedding)
        similarities.append((similarity, documents[idx]))

    # sort by similarity
    similarities.sort(reverse=True)
    return similarities[:top_k]
```

```python
results = semantic_search(query, sentences, embeddings, 2)
results
```
