# **Open AI Embeddings**

```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain_openai import OpenAIEmbeddings


embeddings = OpenAIEmbeddings(
    base_url="BASE_URL", 
    model="text-embedding-3-large"
)
```

```python
## create your first embeddings
text = "Hello, I am learning about embeddings!"

single_embedding = embeddings.embed_query(text)
print(f"Text: {text}")
print(f"Embedding Length : {len(single_embedding)}")
print(single_embedding)
```

```python
# create embedding for list of documents
sentences = [
    "The quick brown fox jumps over the lazy dog.",
    "Data science drives decisions in modern organizations.",
    "Machine learning models improve with more data.",
    "Embedding vectors capture semantic meaning of text.",
    "Python is a versatile programming language.",
    "AI can assist humans in complex tasks.",
    "Cloud computing provides scalable resources.",
    "Collaborative teams build better software.",
    "Documentation helps users understand code.",
    "Testing ensures reliability of applications."
]

multiple_embedding = embeddings.embed_documents(sentences)
print("Embedded sentences : ")
for idx in range(len(sentences)):
    print(f"{sentences[idx]} : {multiple_embedding[idx]}")
```
